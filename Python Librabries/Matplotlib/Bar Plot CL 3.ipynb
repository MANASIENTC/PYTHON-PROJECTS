# Bar Plot

### Vertical Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]

plt.bar(X,Y)
plt.show()
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]

plt.bar(X,Y)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("WSCUBE")
plt.show()
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]
Z= [20,40,50,65]

plt.xlabel("Languages" ,fontsize = 15 )
plt.ylabel("Popularity",fontsize = 15)
plt.title("WSCUBE",fontsize = 15)
c= ["y","b","m","g"]
plt.bar(X,Y, width = 0.4 , color = c,label = "Popularity")
plt.bar(X,Z, width = 0.4 , color = "Orange",label = "Popularity1")
plt.legend()
plt.show()
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]
Z = [20,40,50,65]
P = np.arange(len(X))

plt.xlabel("Languages" ,fontsize = 15 )
plt.ylabel("Popularity",fontsize = 15)
plt.title("WSCUBE",fontsize = 15)
width = 0.3

c= ["y","b","m","g"]
plt.bar(P,Y, width,color = c,label = "Popularity")
plt.bar(P,Z ,width, color = "Orange",label = "Popularity1")
plt.legend()
plt.show()
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]
Z = [20,40,50,65]

width = 0.3
P = np.arange(len(X))
P1 = [j + width for j in P]

plt.xlabel("Languages" ,fontsize = 15 )
plt.ylabel("Popularity",fontsize = 15)
plt.title("WSCUBE",fontsize = 15)


c= ["y","b","m","g"]
plt.bar(P,Y, width,color = c,label = "Popularity")
plt.bar(P1 ,Z ,width, color = "Orange",label = "Popularity1")

plt.legend()
plt.show()
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]
Z = [20,40,50,65]

width = 0.3
P = np.arange(len(X))
P1 = [j + width for j in P]

plt.xlabel("Languages" ,fontsize = 15 )
plt.ylabel("Popularity",fontsize = 15)
plt.title("WSCUBE",fontsize = 15)


c= ["y","b","m","g"]
plt.bar(P,Y, width,color = c,label = "Popularity")
plt.bar(P1 ,Z ,width, color = "Orange",label = "Popularity1")

plt.xticks(P + width / 2, X, rotation = 20)
plt.legend()
plt.show()
```

### Horizontal Bar Graph

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]
Z= [20,40,50,65]

plt.xlabel("Languages" ,fontsize = 15 )
plt.ylabel("Popularity",fontsize = 15)
plt.title("WSCUBE",fontsize = 15)
c= ["y","b","m","g"]

plt.barh(X,Y, color = "y", label = "Popularity")
plt.barh(X,Z, color = "Orange",label = "Popularity1")

plt.legend()
plt.show()
```

```python
X = ["Python","C","C++","Java"]
Y = [88,70,78,82]
Z = [20,40,50,65]

width = 0.3
P = np.arange(len(X))
P1 = [j + width for j in P]

plt.xlabel("Languages" ,fontsize = 15 )
plt.ylabel("Popularity",fontsize = 15)
plt.title("WSCUBE",fontsize = 15)


c= ["y","b","m","g"]
plt.barh(P,Y, width,color = c,label = "Popularity")
plt.barh(P1 ,Z ,width, color = "Orange",label = "Popularity1")

plt.xticks(P + width / 2, X, rotation = 20)
plt.legend()
plt.show()
```
