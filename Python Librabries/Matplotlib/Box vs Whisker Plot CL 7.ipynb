# Box vs Whisker Plot

```python
import matplotlib.pyplot as plt
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X)
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, notch = True)
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X,vert = True)
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, vert = False)
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, widths = 0.2)           # Change width(0.2 is default width)
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, widths = 0.5)           # Change width(0.2 is default width)
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, labels= ["Python"])           # Give label.
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, labels = ["Python"], patch_artist= True)           # To fill box with color by using patch.
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, labels = ["Python"],showmeans= True )           # By using showmeans we display the mean.
plt.show()
```

```python
X = [10,20,30,40,50,60,70,110,200]

plt.boxplot(X, labels = ["Python"],showmeans= True )         # To create outlier.   
plt.show()
```

```python
X = [10,20,30,40,50,60,70,110,200]

plt.boxplot(X, labels = ["Python"],showmeans= True, whis= 3.5 )         # Using whis join the line with outlier.   
plt.show()
```

```python
X = [10,20,30,40,50,60,70,110,200]

plt.boxplot(X, labels = ["Python"],showmeans= True, sym ="g+" )         # By using "sys" to change the shape of outlier and color.    
plt.show()
```

```python
X = [10,20,30,40,50,60,70,110,200]

plt.boxplot(X, labels = ["Python"],showmeans= True,  sym = " " )         # To hide the outlier.   
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, labels = ["Python"],showmeans= True,
            boxprops = dict(color="r"),capprops=dict(color= "r") )         # To change the color of  caps.   
plt.show()
```

```python
X = [10,20,30,40,50,60,70]

plt.boxplot(X, labels = ["Python"],showmeans= True, 
            boxprops = dict(color="r"),capprops=dict(color= "r"), whiskerprops= dict(color= "r"))         # To change the color of  caps middle lines.   
plt.show()
```

```python
X = [10,20,30,40,50,60,70,120]

plt.boxplot(X, labels = ["Python"],showmeans= True,sym="g+", boxprops = dict(color="r"),
            capprops=dict(color= "r"), whiskerprops= dict(color= "r"),flierprops=dict(color="y"))         # To change the color of  caps middle lines.   
plt.show()
```

```python
X = [10,20,30,40,50,60,70]
X1 = [10,20,40,50,30,60,90]
Y = [X, X1]

plt.boxplot(Y,labels =["Python", "C++"],showmeans= True,sym="g+", boxprops = dict(color="r"),
            capprops=dict(color= "r"), whiskerprops= dict(color= "r"),flierprops=dict(color="y"))         # To create multiple boxplot.   


plt.show()
```
