## Pie Plot

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]

plt.pie(X)

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]

plt.pie(X, labels = Y)               # Add labels

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]

plt.pie(X, labels = Y, explode = EX)           # Explode the  the given data

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C)         # Give color name

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%")        # Add percentage

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, autopct = "%0.2f%%")        # Add percentage

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%", shadow = True)        # Add Shadow 

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%", shadow = True, radius= 0.7)        # To give Radius  

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%", shadow = True, radius= 0.7, labeldistance = 1.3)        # To give Label Distance  

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%", shadow = True, radius= 1.0, labeldistance = 0.8)        # To give label Distance  

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%", shadow = True, radius= 1.0, labeldistance = 0.8, startangle = 90)        # To give starting angle to the circle.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX, colors =C, autopct = "%0.1f%%", shadow = True, radius= 1.0, labeldistance = 1.2, startangle = 180)        # To give starting angle to the circle.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15})        # Give the fontsize to Text in the Data.
plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= True)    # Set the labels as anticlock wise direction.(It's a default)

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False)        #Set the labels as a clock wise direction.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5})                 # To increse the width of shadow.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5, "edgecolor": "m"})                 # Give the color to the edges.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5, "edgecolor": "m"}, center = (1,3))                 # To chaange the center.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5}, rotatelabels = True)      # Rotate the labels.

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5}, rotatelabels = False)      # Rotate the labels. (It is  Default)

plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5}, rotatelabels = False)      

plt.title("Course")             # Add Title.
plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5}, rotatelabels = False)      

plt.title("Course")       
plt.legend()                 # Add legend (for which color indicate the Course name)
plt.show()
```

```python
X = [10,20,30,40]
Y = ["C", "C++","Java","Python"]
EX= [0.3,0.0,0.0,0.0]
C = ["r","b","g","y"]

plt.pie(X, labels = Y, explode = EX,  autopct = "%0.1f%%", shadow = True, radius= 1.0, 
        labeldistance = 1.2, startangle = 0, textprops = {"fontsize": 15}, counterclock= False,
        wedgeprops = {"linewidth":5}, rotatelabels = False)      

plt.title("Course")       
plt.legend(loc = 2)                 # Change location using "loc"
plt.show()
```

```python
# To create multiple pie chart in one diagram.

X = [10,20,30,40]
X1 = [40,30,20,10]
Y = ["C", "C++","Java","Python"]

C = ["r","b","g","y"]

plt.pie(X, labels = Y, radius = 1.5)      
plt.pie(X1,radius = 0.8, colors = C)
      
plt.show()             
```

```python
# To create ring pie chart.

X = [10,20,30,40]
X1 = [40,30,20,10]
Y = ["C", "C++","Java","Python"]

C = ["r","b","g","y"]

plt.pie(X, labels = Y, radius = 1.5)      
plt.pie([1], colors = "w")  
      
plt.show()               
```

```python
# To create ring pie chart.

X = [10,20,30,40]
X1 = [40,30,20,10]
Y = ["C", "C++","Java","Python"]

C = ["r","b","g","y"]

plt.pie(X, labels = Y, radius = 1.5)      
cr = plt.Circle(xy=(0,0), radius = 1,  facecolor ="w")
plt.gca().add_artist(cr)
      
plt.show()               
```
