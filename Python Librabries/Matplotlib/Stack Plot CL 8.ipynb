# Stack Plot

```python
# Stack plot also called as "Area Plot". We where plot multiple plot in one is called as Stack Plot.
# It represent the area.
```

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Area = [1,2,3,4,5]

plt.stackplot(X, Area)
plt.show()
```

```python
X = [1,2,3,4,5]
Area = [2,3,2,5,4]

plt.stackplot(X, Area)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,6,4,5,2]

plt.stackplot(X, Area1,Area2,Area3)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l)             # Give the labels.
plt.legend()
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l)           # Givee the labels with location.
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"])   # Add the colors for the labels.
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"], baseline = "zero")     # Adjust the baseline.
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"], baseline = "sym")       # Adjust the baseline.
plt.legend(loc =2) 
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"], baseline = "wiggle")      # Adjust the baseline.
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"])

plt.title("Python")                 # Add the title.
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"])

plt.title("Python")
plt.xlabel("x-axis")                 # Add the X axis and Y axis.
plt.ylabel("y-axis")          
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"])

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis") 
plt.grid()                          # Add Grid.
plt.legend(loc =2)
plt.show()
```

```python
X = [1,2,3,4,5]
Area1 = [2,3,2,5,4]
Area2 = [3,1,2,3,7]
Area3 = [1,4,5,2,6]

l = ["area1","area2","area3"]

plt.stackplot(X, Area1,Area2,Area3, labels=l, colors = ["y","r","g"])

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")          
plt.legend(loc =2)
plt.show()
```
