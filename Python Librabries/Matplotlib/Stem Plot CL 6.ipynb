## Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y)          
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt =":")                 # Using Dotted line
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt ="--")                 # Using Dashed line
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt ="-.")                 # Using Dash-Dot Line
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt = ":", markerfmt = "r+" )                 # To change marker with color.
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt = ":", markerfmt = "ro", bottom = 9 )                 # To change bottom line.
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt = ":", markerfmt = "ro", bottom =0, basefmt = "g" )                 # To change base line color..
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt = ":", markerfmt = "ro", bottom =0, basefmt = "g", label  ="Python" )               # Give the labels.
plt.legend()
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt = ":", markerfmt = "ro", bottom =0, basefmt = "g", 
                    label  ="Python", orientation = "horizontal" )               #Horizontal Graph
plt.legend()
plt.show()
```

```python
X = [1,2,3,4,5,6]
Y = [2,2,5,6,4,3]

plt.stem(X,Y, linefmt = ":", markerfmt = "ro", bottom =0, basefmt = "g", 
                    label  ="Python", orientation = "vertical" )             # Vertical Grapg(It is By default)
plt.legend()
plt.show()
```
