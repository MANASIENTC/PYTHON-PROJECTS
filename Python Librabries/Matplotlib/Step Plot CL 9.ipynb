# Step Plot

```python
# Step Plot : It show the graph with step by step.
```

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Y = [1,2,3,4,5]

plt.step(X, Y)
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y)
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y)
plt.grid()            # Add Grid
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y)

plt.title("Python")            # Add title, x axis and y axis.
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.grid()
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y, marker = "o")        # Add marker 

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.grid()
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y,color = "r", marker = "o")     # Change color.

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.grid()
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y,color = "r", marker = "o",ms = 10)       

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.grid()
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y,color = "r", marker = "o", ms=10, mfc= "black")        # Change marker color.

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.grid()
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y,color = "r", marker = "o", ms=10, mfc= "black", label = "Python")      # Add label

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.legend()
plt.grid()
plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,2,3,5,4]

plt.step(X, Y,color = "r", marker = "o", ms=10, mfc= "black", label = "Python")

plt.title("Python")
plt.xlabel("x - axis")
plt.ylabel("y - axis")

plt.legend(loc =2)                    # Change location of label
plt.grid()
plt.show()
```
