# Matplotlib Subplot Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y, color="r")
plt.show()   
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y, color="r")
plt.pie([1], colors ="y")
plt.show() 
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y, color="r")
plt.pie([1], colors ="y")

x1 = [10,20,30,40]
plt.pie(x)

x2 = ["a","s","d","f"]
y2 = [10,20,30,40]
plt.bar(x2,y2)

plt.show() 
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y, color ="r")

plt.subplot(2,2,2)
plt.pie([1], colors ="y")

x1 = [10,20,30,40]
plt.subplot(2,2,3)
plt.pie(x)

x2 = ["a","s","d","f"]
y2 = [10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.show() 
```
