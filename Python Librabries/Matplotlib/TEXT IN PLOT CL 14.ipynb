# Text in Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)
plt.title("GRAPH")   # Give title

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.title("GRAPH", fontsize= 15)   # Change fontsize of Title

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.title("GRAPH", fontsize= 15)   

plt.xlabel("Days", fontsize =15)    
plt.ylabel("Python", fontsize =15)     # Give y-label and fontsize.

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.title("GRAPH", fontsize= 15)   

plt.xlabel("Days", fontsize =15)    
plt.ylabel("Python", fontsize =15)   

plt.text(2,5,"java",fontsize =15)       # Change fontsize.

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.title("GRAPH", fontsize= 15)   

plt.xlabel("Days", fontsize =15)    
plt.ylabel("Python", fontsize =15)   

plt.text(2,5,"java",fontsize =15, style="italic")       # Add Style

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.title("GRAPH", fontsize= 15)   

plt.xlabel("Days", fontsize =15)    
plt.ylabel("Python", fontsize =15)   

plt.text(2,5,"java",fontsize =15, style="italic", bbox = {"facecolor" : "r"})       # Add facecolor

plt.show()
```

```python
X = [1,2,3,4,5]
Y = [3,1,5,2,6]

plt.plot(X,Y)

plt.title("GRAPH", fontsize= 15)   

plt.xlabel("Days", fontsize =15)    
plt.ylabel("Python", fontsize =15)   

plt.text(2,5,"java",fontsize =15, style="italic", bbox = {"facecolor" : "r"})       
plt.annotate("python", xy= (3,2),xytext=(4,4),
            arrowprops=dict(facecolor="black", shrink=100))            # Add annotate

plt.show()
```
