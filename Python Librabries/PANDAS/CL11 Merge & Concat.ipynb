# MERGE

```python
import pandas as pd

df= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
df1= pd.DataFrame({"A": [1,2,3,5],"C":[21,22,23,24]})

pd.merge(df1,df,on="A")
```

```python
pd.merge(df,df1, how= "inner")
```

```python
pd.merge(df,df1, how="left")
```

```python
pd.merge(df,df1, how="left")
```

```python
pd.merge(df,df1, how="outer")
```

```python
pd.merge(df,df1, how="outer", indicator = True)
```

```python

df= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
df1= pd.DataFrame({"A": [1,2,3,5],"B":[21,22,23,24]})
pd.merge(df,df1)
```

```python
pd.merge(df,df1, left_index = True, right_index = True)
```

```python
pd.merge(df,df1, left_index = True, right_index = True, suffixes = ("Name","ID"))
```

# CONCAT

```python
sr = pd.Series([1,2,3,4])
sr1= pd.Series([11,21,31,41])
pd.concat([sr,sr1])
```

```python
d1= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
d2= pd.DataFrame({"A": [1,2,3,5],"B":[21,22,23,24]})

pd.concat([d1,d2])
```

```python
pd.concat([d1,d2], axis = 1)
```

```python
pd.concat([d1,d2], axis = 0)
```

```python
d1= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
d2= pd.DataFrame({"A": [1,2],"C":[21,22]})

pd.concat([d1,d2],axis = 1)
```

```python
d1= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
d2= pd.DataFrame({"A": [1,2],"C":[21,22]})

pd.concat([d1,d2],axis = 1, join= "outer")
```

```python
d1= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
d2= pd.DataFrame({"A": [1,2],"C":[21,22]})

pd.concat([d1,d2],axis = 1,join= "inner")
```

```python
d1= pd.DataFrame({"A": [1,2,3,4],"B":[11,12,13,14]})
d2= pd.DataFrame({"A": [1,2,3,4],"C":[21,22,23,24]})

pd.concat([d1,d2],axis = 0 , keys= ["d1", "d2"]  )
```
