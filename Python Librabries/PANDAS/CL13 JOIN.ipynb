# JOIN

```python
import pandas as pd
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]})
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]})

df.join(df1)
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df.join(df1)
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df1.join(df)
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df.join(df1,  how= "left")
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df.join(df1,  how = "right")
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df1.join(df, how = "outer")
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df.join(df1, how= "inner")
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]}, index= ["a","b","c","d"])
df1 =pd.DataFrame({"C" : [10,20],"D" :[11,22]}, index= ["a","b"])

df1.join(df, how= "outer")
```

```python
df = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]})
df1 =pd.DataFrame({"C" : [10,20],"B" :[11,22]}, index= ["a","b"])

df1.join(df, how= "outer", lsuffix = "_12", rsuffix = "_12")
```

# Append

```python
var1 = pd.DataFrame({"A" : [1,2,3,4],"B" :[11,12,13,14]})
var2 = pd.DataFrame({"C" : [10,20],"B" :[11,22]})

var1.append(var2)
```
