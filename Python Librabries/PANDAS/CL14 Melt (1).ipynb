# Melt()

```python
import pandas as pd
```

```python
df= pd.DataFrame({"day":[1,2,3,4,5,6],"eng":[10,12,14,15,16,12],"maths": [17,18,19,13,14,16]})
df
```

```python
pd.melt(df)           # Using melt all data show as verticle reshape
```

```python
pd.melt(df, id_vars = ["eng"])            
```

```python
pd.melt(df, id_vars = ["day"] )  
```

```python
pd.melt(df, id_vars = ["day"], var_name="Python" ) 
```

```python
pd.melt(df, id_vars = ["day"], var_name="Python",value_name = "Wsc" ) 
```
