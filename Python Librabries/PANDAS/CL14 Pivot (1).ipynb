# Pivot ()

```python
import pandas as pd

df= pd.DataFrame({"day":[1,2,3,4,5,6],"st_name": ["a","b","a","a","b","b"],
                  "eng":[10,12,14,15,16,12],"maths": [17,18,19,13,14,16]})
df
```

```python
# Using this Pivot() we seperate perticular columns data.
```

```python
df.pivot(index = "day",columns = "st_name")
```

```python
df.pivot(index = "day",columns = "st_name",values = "eng")
```

```python
df= pd.DataFrame({"day":[1,1,1,1,2,2],"st_name": ["a","b","a","a","b","b"],
                  "eng":[10,12,14,15,16,12],"maths": [17,18,19,13,14,16]})
df
```

```python
df.pivot_table(index = "st_name", columns = "day", aggfunc = "mean")
```

```python
df.pivot_table(index = "st_name", columns = "day", aggfunc = "sum")
```

```python
df.pivot_table(index = "st_name", columns = "day", aggfunc = "count")
```

```python
df.pivot_table(index = "st_name", columns = "day",
               aggfunc = "sum", margins = "True")
```

```python
df.pivot_table(index = "st_name", columns = "day",
               aggfunc = "sum", margins = "True")
```
