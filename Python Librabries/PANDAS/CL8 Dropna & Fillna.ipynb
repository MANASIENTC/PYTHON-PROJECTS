# Dropna & Fillna () 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\manas\Downloads\NFL Play by Play 2009-2016 (v3).csv")
df
```

```python
var = df.head(20)
var
```

```python
var.dropna(axis = 1)  # Drop Axis 
```

```python
var.dropna(how ="any" )          # Remove NA row
```

```python
var.dropna(how="all")            # All value get
```

```python
var.dropna(subset=["yacEPA"])    # Using "subset" will remove the Null values from the "columns"
```

var.dropna(inplace = True)               # Nulll value remove and replace new data set
var                               

```python
var.dropna(thresh = 1)
```

# FIILLNA()

```python
var.fillna("Python")
```

```python
var
```

```python
var.fillna({"GameID ":"Python", "yacEPA":"Java", "airWPA": "ML"})  # To fill perticular value in column using this
```

```python
var.fillna(method = "Bfill")    # To fill Null value using Forward value using "F" and "B" for Backward value.
```

```python
var.fillna(method = "bfill", axis = 1)   # To fill null value along with axis
```

```python
var.fillna("Python", limit=10)          # To fill value in null value in limit some perticilar value in range in column
```
